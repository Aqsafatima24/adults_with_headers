### pre-process

```python
# One-hot encoding for 'FuelType'
df = pd.get_dummies(df, columns=['Fuel_Type'], drop_first=True)
df
```

```python
# Normalize or standardize numerical variables if needed
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Age_08_04', 'KM', 'HP', 'cc', 'Weight']] = scaler.fit_transform(df[['Age_08_04', 'KM', 'HP', 'cc', 'Weight']])
df
```

### Split the dataset into training and testing sets (e.g., 80% training, 20% testing).

```python
from sklearn.model_selection import train_test_split

# Define independent variables (X) and dependent variable (y)
X = df.drop('Price', axis=1)  # Independent variables
y = df['Price']  # Dependent variable

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=48)
```

### MODELS

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Predict on test set
y_pred1 = model1.predict(X_test)

# Model evaluation
print("R-squared:", r2_score(y_test, y_pred1).round(3))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred1).round(3)))
```

```python
X_train2 = X_train.drop(['HP'], axis=1)
X_test2 = X_test.drop(['HP'], axis=1)

# Train model
model2 = LinearRegression()
model2.fit(X_train2, y_train)

# Predict and evaluate
y_pred2 = model2.predict(X_test2)
print("R-squared:", r2_score(y_test, y_pred2).round(3))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred2).round(3)))
```

### 1. What is Normalization & Standardization and how is it helpful?
    
**Normalization** scales data to a range, typically between 0 and 1. Itâ€™s useful when the data has varying scales and you want all features to contribute equally in machine learning algorithms.

**Standardization** transforms data to have a mean of 0 and a standard deviation of 1. Itâ€™s helpful when the data is normally distributed or when the model assumes a normal distribution.

### 2.What techniques can be used to address multicollinearity in multiple linear regression?

Techniques to address multicollinearity in multiple linear regression:
1. Remove highly correlated variables: If two variables are highly correlated, consider removing one.
2. Combine variables: Create a new variable that combines the two correlated variables (e.g., average or sum).
3. Principal Component Analysis (PCA): Reduces dimensionality by creating uncorrelated components from the original variables.
4. Regularization (Lasso or Ridge): These techniques penalize large coefficients, helping to reduce the impact of multicollinearity.
5. Variance Inflation Factor (VIF): Calculate VIF for each variable and remove those with a high VIF, indicating multicollinearity.
